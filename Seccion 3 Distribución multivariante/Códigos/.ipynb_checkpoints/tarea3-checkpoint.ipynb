```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
datos = pd.read_csv('../Datos/companies79.csv')
```

```python
datos.head()
```

```python
xf = datos.iloc[:,2:]
xf.head()
```

```python
sns.pairplot(xf, hue = 'V8', diag_kind='kde', diag_kws=dict(shade = True), height=1.5)
plt.show()
```

```python
var = ['V2', 'V3', 'V4', 'V5', 'V6', 'V7']
xf[var] = xf[var].applymap(lambda x: np.log10(x+1))
```

```python
xf.head()
```

```python
sns.pairplot(xf, hue = 'V8', diag_kind='kde', diag_kws=dict(shade = True), height=1.5)
plt.show()
```
